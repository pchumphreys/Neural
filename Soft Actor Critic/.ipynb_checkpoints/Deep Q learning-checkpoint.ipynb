```python
%load_ext autoreload
% autoreload 2
import dqn_expm
import numpy as np
import tensorflow as tf
from dqn_expm_params import expm_params

expm_name = 'CartPole'
params = expm_params[expm_name]

learning_rates = [1e-4,1e-3]
runs_per_point = 3
max_episodes = params['runner']['max_episodes']

rewards = np.zeros((len(learning_rates),runs_per_point,max_episodes))
for i,lr in enumerate(learning_rates):
    params['agent']['learning_rate']= lr            
    for j in range(runs_per_point):
        runner = dqn_expm.run_expm(expm_name, params = params)
        rewards[i,j] = runner.episode_rewards
```

```python
from matplotlib import pyplot as plt

x = range(np.shape(rewards)[-1])
ys = np.mean(rewards,axis=1)
errors = np.std(rewards,axis=1)

for y,error,lr in zip(ys,errors,learning_rates):
    plt.plot(x, y, '-',label=lr)
    plt.fill_between(x, y-error, y+error,alpha=0.2)
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.legend()
plt.show()
plt.close()
```
